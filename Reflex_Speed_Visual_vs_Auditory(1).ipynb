```python
import matplotlib.pyplot as plt
```

```python
VIS1_MEAN = 410.5
VIS2_MEAN = 501.5
VIS1_STD = 91.63
VIS2_STD = 191.05
```

```python
# build a bar plot
plt.bar(['Person 1 Visual', 'Person 2 Visual'],[VIS1_MEAN, VIS2_MEAN], yerr=[VIS1_STD, VIS2_STD], capsize=10)
plt.ylabel('Speed (msec)')
plt.xlabel("Person")
plt.title('Reflex Speed, Visual')
```
